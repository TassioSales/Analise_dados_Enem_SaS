```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

<h1 text-align: center>A estratégia de imputação.</h1>

<p>Se "mean", substitua os valores ausentes usando a média ao longo de cada coluna. Só pode ser usado com dados numéricos.</p>

<p>Se "median", substitua os valores ausentes usando a mediana ao longo de cada coluna. Só pode ser usado com dados numéricos.</p>

<p>Se "most_frequent", substitua a falta usando o valor mais frequente ao longo de cada coluna. Pode ser usado com strings ou dados numéricos. Se houver mais de um valor desse valor, apenas o menor é devolvido.</p>

<p>Se "constant", substitua os valores perdidos por fill_value. Pode ser usado com strings ou dados numéricos.</p>

```python
quantitativas = SimpleImputer(missing_values=np.nan, strategy='mean')
quantitativas_2 = SimpleImputer(missing_values=np.nan, strategy='median') 
qualitativas = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
df = pd.read_excel("Tassio'_'ENEM_DADOS_Anomalias.xlsx")
```

```python
display(df)
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
#Variaveis quantitativas
df['NU_NOTA_MT'] = quantitativas_2.fit_transform(df[['NU_NOTA_MT']])
df['NU_NOTA_LC'] = quantitativas_2.fit_transform(df[['NU_NOTA_LC']])
df['NU_NOTA_CN'] = quantitativas_2.fit_transform(df[['NU_NOTA_CN']])
df['NU_NOTA_CH'] = quantitativas_2.fit_transform(df[['NU_NOTA_CH']])
df['NU_NOTA_REDACAO'] = quantitativas_2.fit_transform(df[['NU_NOTA_REDACAO']])
df['NOTA_MEDIA'] = quantitativas_2.fit_transform(df[['NOTA_MEDIA']])
df['NU_IDADE'] = quantitativas_2.fit_transform(df[['NU_IDADE']])
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
#Variaveis qualitativas
df['Sexo'] = qualitativas.fit_transform(df[['Sexo']])
df['COR_RACA'] = qualitativas.fit_transform(df[['COR_RACA']])
df['Tipo_Escola'] = qualitativas.fit_transform(df[['Tipo_Escola']])
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
#Lembrando que vc pode salvar o arquivo tanto em Csv como XLS

df.to_csv('pessoas.csv', sep = ";", encoding = "utf-8", index=False)
```
